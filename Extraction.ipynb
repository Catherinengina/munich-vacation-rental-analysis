```python
import os

folders = [
    "data/raw",
    "data/processed",
    "outputs"
]

for folder in folders:
    os.makedirs(folder, exist_ok=True)
    print(f"✅ Created: {folder}")

print("\nFolder structure ready!")
```

```python
import gzip
import shutil
from pathlib import Path

DATA_RAW = Path("data/raw")

gz_files = list(DATA_RAW.glob("*.gz"))

for gz_path in gz_files:
    out = gz_path.with_suffix("")  # removes .gz
    if out.exists():
        print(f"  [SKIP] {out.name} already decompressed.")
        continue
    print(f"  Decompressing {gz_path.name} ...")
    with gzip.open(gz_path, "rb") as fi, open(out, "wb") as fo:
        shutil.copyfileobj(fi, fo)
    print(f"  ✅ Done: {out.name}")

print("\n✅ All files decompressed!")
```
